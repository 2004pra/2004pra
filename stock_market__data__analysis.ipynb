```python
df=df[['S.No.', 'name', 'Sales_ Growth_5yrs', 'Profit_Growth_5yrs', 'ROE_5Yrs',
       'Stock_Price_Return_5yrs', 'Free _Cash_ Flow _5yrs', 'M.Cap/Sales',
       'Debt/Profit',
    #'overall_rank',
    'Score']]
```

```python
df['Score'] = (
    df['Profit_Growth_5yrs'].rank(ascending=False) +
    df['ROE_5Yrs'].rank(ascending=False) +
    df['Stock_Price_Return_5yrs'].rank(ascending=False)
)
```

```python
df.columns
```

```python
top_10_companies = df.nsmallest(10, 'Score')['name']
```

```python
df['Top_10'] = df['name'].isin(top_10_companies)
```

```python
selected_cols = ['Profit_Growth_5yrs', 'ROE_5Yrs', 'Stock_Price_Return_5yrs', 'Top_10']
```

```python
sns.pairplot(df[selected_cols], hue='Top_10', palette={True: 'red', False: 'blue'})
plt.show()
```
